```python
import pandas as pd
```

```python
df = pd.read_csv("human_vital_signs_dataset_2024.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.drop( columns=['Patient ID', 'Timestamp'], axis = 1, inplace=True)
```

```python
df['Gender'] = df['Gender'].replace({
    'Male': 0,
    'Female': 1
})

df['Risk Category'] = df['Risk Category'].replace({
    'Low Risk': 0,
    'High Risk': 1
})
```

```python
df.info()
```

```python
X = df.drop('Risk Category', axis = 1)
y = df['Risk Category']
```

```python
y.value_counts()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(X.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')   # 1 neuron for binary classification
])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

history = model.fit(
    X, y,
    validation_split=0.15,
    epochs=30,
    batch_size=32,
    verbose=1
)
```

```python
df.columns
```

```python
model.save("my_vitals_model.h5")
```

```python
import tensorflow as tf

model = tf.keras.models.load_model("my_vitals_model.h5")

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("vitals_model.tflite", "wb") as f:
    f.write(tflite_model)
```
